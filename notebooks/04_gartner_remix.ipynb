# 04 - Gartner Hype Cycle and the new double Death Valley

Two visualizations from the paper's Section 6.5:
1. The classical Gartner Hype Cycle versus a reformulated double-valley curve for the post-AI regime.
2. The classical death valley (single trough) versus a double-valley cash trajectory in which the *commoditization valley* appears post-product-market-fit.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

from src.hype_cycle import classical_hype_curve, post_genai_double_valley_curve
from src.death_valley import classical_cash_trajectory, post_genai_double_valley_cash_trajectory, detect_valleys
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Hype cycle
n = 32
classical = classical_hype_curve(n_quarters=n)
post = post_genai_double_valley_curve(n_quarters=n)
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(np.arange(n + 1), classical, label='Classical', color='#3498db', linewidth=2.2)
ax.plot(np.arange(n + 1), post, label='Post-GenAI (double valley)', color='#e74c3c', linewidth=2.2)
ax.set_xlabel('Quarters'); ax.set_ylabel('Expectations')
ax.set_title('Hype Cycle - classical vs post-AI')
ax.legend(); plt.show()
```

```python
# Death valley
classical_cash = classical_cash_trajectory()
post_cash = post_genai_double_valley_cash_trajectory()
valleys = detect_valleys(post_cash)
print(f'Valleys detected in post-AI trajectory: {valleys}')

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(classical_cash / 1e6, label='Classical', color='#3498db', linewidth=2.2)
ax.plot(post_cash / 1e6, label='Post-GenAI', color='#e74c3c', linewidth=2.2)
ax.axhline(0, color='black', linewidth=0.5)
ax.set_xlabel('Months'); ax.set_ylabel('Cash (USD M)')
ax.set_title('Death Valley - classical vs post-AI'); ax.legend()
plt.show()
```
